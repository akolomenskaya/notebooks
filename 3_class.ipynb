```
import pandas as pd
import catboost
```

```
train = pd.read_csv('/home/train.csv').fillna(-1)
columns = list(train.columns)
columns.remove('PassengerId')
columns.remove('Survived')
target = train.Survived.values
train = train[columns]
print(len(train))
```

```
categorical = [
    list(train.columns).index(column) for column in train.columns if train[column].dtype == 'object'
]

print(categorical)
```

```
split = len(train) // 2
classifier = catboost.CatBoostClassifier(random_state=0).fit(train[:split], target[:split], categorical)
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(target[split:], classifier.predict(train[split:])))
```

```
print(classifier.score(train[split:], target[split:]))
```
